   # WATERMARKING IMAGES USING   openCV

DIGITAL WATERMARKING:-

   The digital watermarking or watermarking explains the ways and mechanisms to hide the data and the data can be a number or text, in digital media, it may be a picture or video. The watermarking is a message that can be embedded into the digital data like video, pictures, and text and the embedded data can be extracted later. 
   The Indian currency is a good example of watermarking and in the general watermarking procedure the genuine image undergoes the embedding procedure along with the watermark and the output generated will be a watermarked image. 


    # IMPORTING LIBRARIES

```python
import cv2
import os
import glob
```

    # GETTING THE WATERMARK

```python
logo=cv2.imread("E:/images/logo.png")
h_logo, w_logo, _ = logo.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(logo)
```

        # GETTING THE FOLDER

once we have the logo, we have to get the path of all the images where we want to place the watermark.
To do this we will use the glob library.

```python
images_path = glob.glob("E:/images/pics/*.*")
```

we loop through all the images and  we load the image, and  we get height and width of the image.

It depends where we would like to place the watermark, we should now choose the location.

It is possible to place it anywhere on the image, but as I decided to place it in the center, weâ€™re going to get the center location of the image.

Using the center coordinates of the image and the with and height of the logo, we cut on the center of the image a portion with the size of the logo. We will call this ROI, (region of interest).

We will then add the ROI with the logo using the cv2.addWeighted(). This is the core of Watermark effect. Using addweighted we can choose to give the opacity to the logo so that it will look nice and smooth when placed over the image.

```python
print("Adding watermark...........")
for img_path in images_path:
    img = cv2.imread(img_path)
    h_img, w_img, _ = img.shape
    center_y = int(h_img / 2)
    center_x = int(w_img / 2)
    top_y = center_y - int(h_logo / 2)
    left_x = center_x - int(w_logo / 2)
    bottom_y = top_y + h_logo
    right_x = left_x + w_logo
    
# Getting ROI (Region Of Interest)
    
    roi = img[top_y: bottom_y, left_x: right_x]

    
# Add the Logo to the Roi
    
    
    result = cv2.addWeighted(roi, 1 ,logo, 0.3, 0)
    
  

        
# Replace the ROI on the image
    
    img[top_y: bottom_y, left_x: right_x] = result
    

# Get filename and save the image

    
    filename = os.path.basename(img_path)
    cv2.imwrite("E:/images/watermarked/watermarked_" + filename, img)
    print("THANK YOU...watermarked copy stored in the desired folder")
        
   
```

Finally displaying the sample image

```python
print("\n\n\n\tsample OUTPUT IMAGE IS....")
plt.imshow(img)
```
